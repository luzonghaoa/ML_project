```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('Combined_News_DJIA.csv')
```

```python
data.head()
```

```python
train = data[data['Date'] < '2015-01-01']
test = data[data['Date'] > '2014-12-31']
#train.tail()
#train = data[data['Date'] < '20150101']
#test = data[data['Date'] > '20141231']
train.tail()
#print(train.shape)
```

```python
example = train.iloc[3,10]
example = example.lower()
example = CountVectorizer().build_tokenizer()(example)
print(example)
```

```python
pd.DataFrame([[x,example.count(x)] for x in set(example)], columns = ['Word', 'Count'])
```

```python
trainheadlines = []
for row in range(0,len(train.index)):
    trainheadlines.append(' '.join(str(x) for x in train.iloc[row,2:27]))
print(trainheadlines[0])
```

```python
basicvectorizer = CountVectorizer()
basictrain = basicvectorizer.fit_transform(trainheadlines)
print(basictrain.shape)
```

```python
basicmodel = LogisticRegression(solver='lbfgs', max_iter=10000)
basicmodel = basicmodel.fit(basictrain, train["Label"])
```

```python
testheadlines = []
for row in range(0,len(test.index)):
    testheadlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
basictest = basicvectorizer.transform(testheadlines)
print(basictest.shape)
```

```python
predictions = basicmodel.predict(basictest)
```

```python
pd.crosstab(test["Label"], predictions, rownames=["Actual"], colnames=["Predicted"])
```

```python
basicwords = basicvectorizer.get_feature_names()
basiccoeffs = basicmodel.coef_.tolist()[0]
coeffdf = pd.DataFrame({'Word' : basicwords, 
                        'Coefficient' : basiccoeffs})
coeffdf = coeffdf.sort_values(['Coefficient', 'Word'], ascending=[0, 1])
coeffdf.head(10)
```

```python
coeffdf.tail(10)
```

```python
advancedvectorizer = CountVectorizer(ngram_range=(2,2))
advancedtrain = advancedvectorizer.fit_transform(trainheadlines)
print(advancedtrain.shape)
```

```python
advancedmodel = LogisticRegression(solver='lbfgs', max_iter=10000)
advancedmodel = advancedmodel.fit(advancedtrain, train["Label"])
```

```python
testheadlines = []
for row in range(0,len(test.index)):
    testheadlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
advancedtest = advancedvectorizer.transform(testheadlines)
advpredictions = advancedmodel.predict(advancedtest)
```

```python
pd.crosstab(test["Label"], advpredictions, rownames=["Actual"], colnames=["Predicted"])
```

```python
slicedData = data.iloc[:,2:27]
slicedData.replace(to_replace="[^a-zA-Z]", value=" ", regex=True, inplace=True)
wholelines = []
for row in range(0,len(slicedData.index)):
    wholelines.append(' '.join(str(x) for x in slicedData.iloc[row,0:25]))
print(len(wholelines))
print(wholelines[25])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import stop_words
```

```python
X_train, X_test, y_train, y_test = train_test_split(wholelines, data['Label'], test_size=.2, random_state=0)
len(y_train)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
basicvectorizer = TfidfVectorizer(ngram_range=(2,3), stop_words='english',max_features=2000)
basictrain = basicvectorizer.fit_transform(X_train)
basictest = basicvectorizer.transform(X_test)
print(basictrain.shape)
print(basictest.shape)
```

```python
basicmodel = LogisticRegression(penalty='l1')
model = basicmodel.fit(basictrain, y_train)
#model.score(test_tfidf, y_test)
predictions = model.predict_proba(basictest)
#print(predictions)
yv = 0.5
predictions_q = []
for i in predictions:
    if i[1] >= yv:
        predictions_q.append(1)
    else:
        predictions_q.append(0)
predictions_q = pd.Series(predictions_q)
y_test = y_test.to_list()
y_test = pd.Series(y_test)
#print(y_test)
pd.crosstab(y_test, predictions_q, rownames=["Actual"], colnames=["Predicted"])
```
